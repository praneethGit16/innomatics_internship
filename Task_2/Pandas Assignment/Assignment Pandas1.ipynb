```python
df=pd.read_excel(r'C:\Users\DELL\Downloads\data.xlsx')
df.head()
                 
```

# To Know the tier of college where computer engineering students got salary more than 5lakhs

```python
df.loc[(df.Specialization=='computer engineering')&(df.Salary>500000),['CollegeTier']]
```

# To retrive the details of Senior Software Engineers who are highly paid

```python
#This analysis gives us the details of senior software who has highest Salary
df.loc[(df.Designation == 'senior software engineer') & (df.Salary==df.Salary.max())]
```

# To retrive data of Females who work from Gurgaon as Systems Analyst

```python
df.loc[(df.Gender=='f')&(df.Designation=='systems analyst'),['JobCity']]
```

# To retrive the data of Software Developers

```python
df.loc[df.Designation== 'software developer',['Specialization']]
```

# To Retrive of Students from ECE who are graduated after 2014

```python
df.loc[(df.Specialization=='electronics and communication engineering')&(df.GraduationYear>=2014)]
```

# To retrive the data of students from AP state who scored above 600 in 'Logical'

```python
df.loc[(df.CollegeState=='Andhra Pradesh')&(df.Logical>=600)]
```
